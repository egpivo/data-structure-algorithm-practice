```python
class BTree:
    def __init__(self, val):
        self.val = val 
        self.left = self.right = None


def solution(root: BTree):
    result = 0
    def dfs(node, expected, current_length):
        nonlocal result
        if not node:
            return 0
        
        if node.val == expected:
            current_length += 1
        else:
            current_length = 1
            
            
        result = max(result, current_length)
        
        dfs(node.left, node.val + 1, current_length)
        dfs(node.right, node.val + 1, current_length)

    dfs(root, root.val, 1)
    return result
        

root = BTree(6) 
root.right = BTree(9) 
root.right.left = BTree(7) 
root.right.right = BTree(10) 
root.right.right.right = BTree(11) 

print(solution(root))
        
        
```

```python
# 5 = 101 --> 2
# 3 = 11  --> 2
# 8 = 1000 -> 1
# 7 = 111 --> 3
# 
# input: positive integer (32-int)
# output: # of ones
# bit mask
#  1, ...,32
#  mask = 2^0 (1), 2^1(10), 2^2...
#  5 & mask --> true -> one; false -> zero


def count_ones(n):
    count = 0

    while n:
        count += n & 1
        n >>= 1
    return count
print(count_ones(5))
print(count_ones(3))
print(count_ones(8))
print(count_ones(7))

# TC: O(log(n))
```

```python
def sort_1_to_n_array(arr):
    i = 0
    while i < len(arr):
        correct_idx = arr[i] - 1
        if arr[i] != arr[correct_idx]:
            arr[i], arr[correct_idx] = arr[correct_idx], arr[i]
        else:
            i += 1
    return arr


```
